### Problem Statement 5: (Find and process Missing Data)
Find and store the percentage of data missing for each column.
Some data can be found and filled in by using information from other cells. Try to find the correct data by using information from other cells and filling it in. Find and store the percentage of data missing for each column.

Hint:
	Population = Male + Female
	Literate = Literate_Male + Literate_Female
	Population  = Young_and_Adult+  Middle_Aged + Senior_Citizen + Age_Not_Stated
	Households = Households_Rural + Households_Urban 

Create a visualization that compares the amount of missing data before and after the data-filling process was done. 


```python
# --PS5
df3=df2.copy()
df3.head()
```

```python
df3['Population']=df3['Population'].fillna(df3['Male']+df3['Female'])
df3['Literate']=df3['Literate'].fillna(df3['Literate_Male'] + df3['Literate_Female'])
df3['Households']=df3['Households'].fillna(df3['Households_Rural']+df3['Households_Urban'])
df3['Population']=df3['Population'].fillna(df3['Young_and_Adult']+ df3['Middle_Aged'] + df3['Senior_Citizen'] +df3['Age_Not_Stated'])
#df3.dropna(inplace=True)
#df3[df3.isnull()]
df3.shape
```

```python
df4=df3.copy()
df4['Female'].fillna(df4['Population']-df4['Male'], inplace=True)

df4['Male'].fillna(df4['Population']-df4['Female'], inplace=True)
df4['Population'].fillna(df4['Male'] + df4['Female'], inplace=True)
df4['Literate'].fillna(df4['Literate_Male'] + df4['Literate_Female'], inplace=True)
#male literate and female literate imputation

df4['Literate_Female'].fillna(df4['Literate'] - df4['Literate_Male'], inplace=True)

df4['Literate_Male'].fillna(df4['Literate'] - df4['Literate_Female'], inplace=True)
#household imputation

df4['Households'].fillna(df4['Households_Rural'] + df4['Households_Urban'], inplace=True)

df4['Households_Rural'].fillna(df4['Households'] - df4['Households_Urban'], inplace=True)

df4['Households_Urban'].fillna(df4['Households'] - df4['Households_Rural'], inplace=True)

#aged imputation

df4['Young_and_Adult'].fillna(df4['Population'] - (df4['Middle_Aged'] + df4['Senior_Citizen'] + df4['Age_Not_Stated']), inplace=True)
df4['Middle_Aged'].fillna(df4['Population'] - (df4['Young_and_Adult'] + df4['Senior_Citizen'] + df4['Age_Not_Stated']), inplace=True)
df4['Senior_Citizen'].fillna(df4['Population'] - (df4['Young_and_Adult'] + df4['Middle_Aged'] + df4['Age_Not_Stated']), inplace=True)
df4['Age_Not_Stated'].fillna(df4['Population'] - (df4['Young_and_Adult'] + df4['Senior_Citizen'] + df4['Middle_Aged']), inplace=True)

#df4.isnull().sum()
df4.shape
df4.head()
```

```python
#Total Number of Dilapidated for rural calculation
data.loc[data["Rural/Urban"] == "Rural", "Total Number of Dilapidated"] = (data["Households_Rural"] * data["Total Number of Dilapidated"])//100
#Total Number of Dilapidated for Urban calculation
data.loc[data["Rural/Urban"] == "Urban", "Total Number of Dilapidated"] = (data["Households_Urban"] * data["Total Number of Dilapidated"])//100
#Total Number of Dilapidated for total calculation
data.loc[data["Rural/Urban"] == "Total", "Total Number of Dilapidated"] = (data["Households"] * data["Total Number of Dilapidated"])//100

```
